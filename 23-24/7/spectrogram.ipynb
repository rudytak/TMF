```python
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile

%matplotlib widget
```

```python
sample_rate, samples = wavfile.read('bluk2.wav')
frequencies, times, spectrogram = signal.spectrogram(samples[:,1], sample_rate, nfft=2**12)

plt.figure(figsize=(8,6))
plt.pcolormesh(times, frequencies, spectrogram, cmap="plasma")
plt.colorbar()
# plt.imshow(spectrogram)
plt.ylim(0, 500)
plt.xlim(5.05, 5.25)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

plt.plot([5.09, 5.185], [80, 300], "w-.", linewidth=2)
plt.show()
```
